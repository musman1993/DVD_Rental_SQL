# PostgreSQL DVD Rental Database - SQL Learning

Query and explore the DVD Rental database using SQLAlchemy and Pandas.


```python
import sqlalchemy
from sqlalchemy import create_engine, text
import polars as pl

# Create engine
engine_dvd = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/dvdrental")

# Test connection
with engine_dvd.connect() as conn:
    result = conn.execute(text("SELECT version()"))
    print("✓ Connected to:", result.scalar())
```

```python
import sqlalchemy
from sqlalchemy import create_engine, text
import polars as pl

# Create engine
engine_northwind = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/northwind")

# Test connection
with engine_northwind.connect() as conn:
    result = conn.execute(text("SELECT version()"))
    print("✓ Connected to:", result.scalar())
```

```python
# Helper function for queries
def query(sql):
    """Execute a SQL query and return results as a Polars DataFrame"""
    with engine_northwind.connect() as conn:
        result = conn.execute(text(sql))
        return pl.from_dicts(result.mappings())

# Example: Query actors
result = query("SELECT * FROM order_details")
result
```
